## №1
Произошло убийство. На месте убийства найдена кровь, которая явно (будем считать, что с вероятностью 1) принадлежит настоящему убийце. Кровь принадлежит редкой группе, которая присутствует всего лишь у 1% населения. И так получилось, что у подсудимого как раз эта редкая группа крови!
    Первым выходит прокурор и говорит: “Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен -- всего 1%; значит, с вероятностью 99% он виновен, предлагаю осудить”. В чём не прав прокурор? Укажите, какие вероятности он оценил, а какие должен был оценить.

### Ответ:

Введем обозначения:
- событие $A$- подсудимый виновен;
- событие $\overline{A}$ - подсудимый невиновен;
- событие $B$ - у человека редкая группа крови;
- событие $\overline{B}$ - у человека обычная, а не редкая группа крови.

Нам дано в условии:
- $P(B) = 0,01$ -  вероятность редкой группы у случайного человека.
- $P(B|A) = 1$ - вероятность редкой группы крови у человека при условии, что он убийца.

~~Будем оценивать непредвзято, и предположим, что вероятность виновности подсудимого 50% в отсутствии других доказательств (скорее всего это неверно, возможно, что чаще обвиняют действительно виновных). Хотя нет, это совсем неправильно :)~~

Лучше посмотрим на статистику убийств: 15 убийств на 100000 населения в 2009 году в России. Предположим, что обвинили случайного человека и сделаем очень консервативную оценку на то, скольких человек в среднем лишает жизни убийца. Пусть будет 1 убийство на 1 убийцу. Вероятность того, что случайный человек убийца и виновен 0,00015 или же 0,015%. То есть, $P(A) = 0,00015$.

Прокурор точно не знает про условные вероятности. Прокурор называет вероятность редкой группы крови у случайного человека - 0,01. И дальше делает вывод, что наш человек с 99% вероятности не случайный, то есть виновный. Не знаю, как еще описать то, что у него в голове творится :)

При этом он оценил $P(\overline{B}) = 1 - P(B) = 1 - 0,01 = 0.99$  - вероятность того, что у человека не редкая группа крови.

Прокурор должен был оценить $P(A|B)$ - вероятность того, что подсудимый виновен, при условии, что у него редкая группа крови. Такая же, как и у настоящего убийцы.

Вспоминаем формулу Байеса:

$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} = \frac{1 \cdot 0,00015}{0,01} = 0,015 = 1,5\%$$

**Вероятность того, что подсудимый виновен, при условии, что у него редкая группа крови всего 1,5%.**

## №2
Затем выходит адвокат, объясняет ошибки прокурора, а сам говорит следующее: “В городе живёт миллион человек. Это значит, что примерно у 10000 из них такая группа крови. Значит, всё, что говорит нам эта кровь -- это то, что подсудимый совершил убийство с вероятностью 0.01%. Получается, что это свидетельство не просто не является строгим доказательством, а и вовсе даёт нам ничтожно малую вероятность того, что мой подзащитный виновен, так что предлагаю исключить его из материалов дела”. В чём не прав адвокат? Укажите, какие вероятности он оценил, а какие должен был оценить.

### Ответ:

Адвокат оценил $P(B)$ - вероятность редкой группы крови у случайного человека, и выдал ее за вероятность виновности подсудимого.

Адвокат должен был оценить $P(\overline{A}|B)$ - вероятность невиновности подсудимого при условии, что у него редкая группа крови.

Очевидно, что $P(\overline{A}|B) = 1 - P(A|B) = 1 - 0,015 = 0,985 = 98,5\%$.

**Вероятность, что подсудимый невиновен, при условии, что у него редкая группа крови 98,5%**

Если вдруг не очевидно, то распишем почему это верно. Запишем формулу Байеса:

$$P(\overline{A}|B) = \frac{P(B|\overline{A}) \cdot P(\overline{A})}{P(B)}$$

Нам нужна вероятность невиновности подсудимого $P(\overline{A})$:

$$P(\overline{A}) = 1 - P(A)$$

Из формулы полной вероятности найдем $P(B|\overline{A})$:

$$P(B) = P(B|A) \cdot P(A) + P(B|\overline{A}) \cdot P(\overline{A})$$

$$P(B|\overline{A}) = \frac{P(B) - P(B|A) \cdot P(A)}{P(\overline{A})} = \frac{P(B) - P(B|A) \cdot P(A)}{1 - P(A)}$$

Подставим $P(\overline{A})$ и $P(B|\overline{A})$ в формулу Байеса:

$$P(\overline{A}|B) = \frac{P(B|\overline{A}) \cdot P(\overline{A})}{P(B)} = \frac{\frac{P(B) - P(B|A) \cdot P(A)}{1 - P(A)} \cdot (1 - P(A))}{P(B)} = \frac{P(B) - P(B|A) \cdot P(A)}{P(B)} = $$

$$ = 1 - \frac{P(B|A) \cdot P(A)}{P(B)} =  1 - P(A|B) $$

## №3
Вернёмся к тесту на коронавирус из лекции. Напомню, что коронавирусом по нашему предположению болеет 1% населения, а тест имеет вероятность ошибки 5% в обе стороны. Мы видели, что в результате человек с положительным тестом имеет апостериорную вероятность оказаться больным всего лишь около 16%.
    Врачи решили, что терпеть это никак нельзя; они готовы провести срочное исследование и улучшить качество теста на коронавирус. Но сил и денег хватит только для того, чтобы сократить один тип ошибок, т.е. уменьшить или только число false positives (когда тест выдаёт позитивный результат у здорового человека), или только число false negatives (когда тест выдаёт отрицательный результат у больного).
    Помогите медикам: постройте зависимости апостериорных вероятностей болезни после положительного и отрицательного теста от обоих видов ошибок (например, в виде графиков), сделайте выводы и дайте рекомендации врачам -- на чём лучше сосредоточиться?

### Решение:

Будем использовать обозначения из лекции:
- d = 0 - человек здоров;
- d = 1 - человек болен;
- t = 0 - тест дал отрицательный результат;
- t = 1 - тест дал положительный результат.


Вероятность болезни при положительном результате теста:

$$P(d = 1|t = 1) = \frac{P(t = 1|d = 1) \cdot P(d = 1)}{P(t = 1)} = $$

$$ = \frac{P(t = 1|d = 1) \cdot P(d = 1)}{P(t = 1|d = 1) \cdot P(d = 1) + P(t = 1|d = 0) \cdot P(d = 0)}$$

Вероятность болезни при отрицательном результате теста:

$$P(d = 1|t = 0) = \frac{P(t = 0|d = 1) \cdot P(d = 1)}{P(t = 0)} = $$

$$ = \frac{P(t = 0|d = 1) \cdot P(d = 1)}{P(t = 0|d = 1) \cdot P(d = 1) + P(t = 0|d = 0) \cdot P(d = 0)}$$

```python
import numpy as np
import matplotlib.pyplot as plt

FIGSIZE = (12,8)
FONTSIZE = 16
PRIOR_PROBA = 0.01
FP_RATES = np.arange(0, 0.0501, 0.001)
FN_RATES = np.arange(0, 0.0501, 0.001)


def compute_posterior(test_is_pos, prior_proba, fp_rate=0.05, fn_rate=0.05):
    if test_is_pos:
        posterior_proba = (
            (1 - fn_rate) * prior_proba /
            ((1 - fn_rate) * prior_proba + fp_rate * (1 - prior_proba))
        )
    else:
        posterior_proba = (
            fp_rate * prior_proba /
            (fn_rate * prior_proba + (1 - fp_rate) * (1 - prior_proba))
        )
    
    return posterior_proba
```

```python
posterior_t1_fixed_fn_rate = compute_posterior(
    True,
    PRIOR_PROBA,
    fp_rate=FP_RATES
)

posterior_t0_fixed_fn_rate = compute_posterior(
    False,
    PRIOR_PROBA,
    fp_rate=FP_RATES
)

plt.figure(figsize=FIGSIZE)
plt.title(
    'Зависимость апостериорной вероятности болезни от false positive rate',
    fontsize=FONTSIZE
)
plt.plot(FP_RATES, posterior_t1_fixed_fn_rate, label='После положительного теста')
plt.plot(FP_RATES, posterior_t0_fixed_fn_rate, label='После отрицательного теста')
plt.legend(fontsize=FONTSIZE)
plt.xlabel('False positive rate', fontsize=FONTSIZE)
plt.ylabel('Вероятность болезни', fontsize=FONTSIZE);
```

```python
posterior_t1_fixed_fp_rate = compute_posterior(
    True,
    PRIOR_PROBA,
    fn_rate=FN_RATES
)

posterior_t0_fixed_fp_rate = compute_posterior(
    False,
    PRIOR_PROBA,
    fn_rate=FN_RATES
)

plt.figure(figsize=FIGSIZE)
plt.title(
    'Зависимость апостериорной вероятности болезни от false negative rate',
    fontsize=FONTSIZE
)
plt.plot(FP_RATES, posterior_t1_fixed_fp_rate, label='После положительного теста')
plt.plot(FP_RATES, posterior_t0_fixed_fp_rate, label='После отрицательного теста')
plt.legend(fontsize=FONTSIZE)
plt.xlabel('False negative rate', fontsize=FONTSIZE)
plt.ylabel('Вероятность болезни', fontsize=FONTSIZE);
```

### Вывод:

По графикам видно, что уменьшение false positive rate позволит выявлять заболевших с большей уверенностью после положительного теста, чем если бы мы уменьшали false negative rate.
